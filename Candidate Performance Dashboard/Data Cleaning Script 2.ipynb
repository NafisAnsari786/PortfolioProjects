```python
import pandas as pd
```

```python
df = pd.read_excel(r"F:\Interview Prep 2025\IMocha\All_Tests_Section_8_13_2024_3_03_21_PM.xlsx")
df.head()
```

```python
# Convert date columns to datetime format
df['Appeared_On'] = pd.to_datetime(df['Appeared_On'], errors='coerce')
```

```python
# Handle missing values (if any)
df.fillna({'Candidate_Status': 'Unknown', 'Candidate_Feedback': 'No Feedback', 'Proctoring_Flag': 'Not Flagged', 'Applicant_ID': 'Unkown'}, inplace=True)
```

```python
# Ensure consistency in column types
df['Candidate_Score'] = pd.to_numeric(df['Candidate_Score'], errors='coerce')
df['Section_Score'] = pd.to_numeric(df['Section_Score'], errors='coerce')
df['Time_Taken(minutes)'] = pd.to_numeric(df['Time_Taken(minutes)'], errors='coerce')
```

```python
# Standardize section names to remove inconsistencies
df['Section_Name'] = df['Section_Name'].str.strip().str.replace("  ", " ")
```

```python
# Aggregate data per candidate and test
summary_df = df.groupby(['Candidate_ID', 'Test_Name']).agg(
    Total_Score=('Candidate_Score', 'sum'),
    Avg_Percentage=('Percentage', 'mean'),
    Total_Time=('Time_Taken(minutes)', 'sum'),
    Sections_Attempted=('Section_Name', 'nunique'),
    Test_Status=('Test_Status', 'first')
).reset_index()
```

```python
# Merge back with main dataset to retain section-level details
final_df = pd.merge(df, summary_df, on=['Candidate_ID', 'Test_Name'], how='left')
```
